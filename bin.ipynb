A. Implement the vector space model with TF-IDF weighting for the
following corpus: [20]

Document 1: "The "The sun is the star at the center ofthe solar system.",

Document2: "She wore a beautiful dress to the party last night."

query = "solar system"

```python
from sklearn.feature_extraction.text import TfidfVectorizer

docs =[
    "The sun is the star at the center of the solar system.",
    "She wore a beautiful dress to the party last night.",
    "The book on the table caught my attention immediately."
]
query = "solar system"

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(docs + [query])
print("TF-IDF Matrix successfully created. Shape:", tfidf_matrix.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1]).flatten()

for i, score in enumerate(similarities):
    print(f"Similarity of Document {i+1} to query: {score:.4f}")
```

A. Implement a learning to rank algorithm (e.g., RankSVM or RankBoost)
on a given dataset. [20]


B. Train the ranking model using labeled data and evaluate its e ffectiveness.

```python
import pandas as pd
from sklearn.ensemble import AdaBoostRegressor

data = {
    'Pclass':[3, 1, 3, 1, 3, 2],
    'Age':[22.0, 38.0, 26.0, 35.0, 35.0, 28.0],
    'Fare':[7.25, 71.28, 7.92, 53.10, 8.05, 13.00],
    'Survived': [0, 1, 1, 1, 0, 0]
}
df = pd.DataFrame(data)

X_train = df[['Pclass', 'Age', 'Fare']]
y_train = df['Survived']

rank_model = AdaBoostRegressor(n_estimators=50, random_state=42)
print("Dataset Loaded. RankBoost (AdaBoost) model initialized successfully.")
```

```python
from sklearn.metrics import ndcg_score
import numpy as np

rank_model.fit(X_train, y_train)

predicted_scores = rank_model.predict(X_train)

print("Predicted Ranking Scores:", predicted_scores)
print("Ranked Passenger Order (Indices):", np.argsort(predicted_scores)[::-1])

ndcg = ndcg_score([y_train.values], [predicted_scores])
print(f"Model Effectiveness (NDCG Score): {ndcg:.4f}")
```

A. Develop a spelling correction module using edit distance algorithms and
find the edit distance between strings "write" and "right".


B. Implement a clustering algorithm (e.g., K-means or hierarchical
clustering). Also, apply the clustering algorithm to a set of documents
and evaluate the clustering results.

```python
import numpy as np

def edit_distance(s1, s2):
    dp = np.zeros((len(s1)+1, len(s2)+1))
    for i in range(len(s1)+1): dp[i, 0] = i
    for j in range(len(s2)+1): dp[0, j] = j
    for i in range(1, len(s1)+1):
        for j in range(1, len(s2)+1):
            cost = 0 if s1[i-1] == s2[j-1] else 1
            dp[i, j] = min(dp[i-1, j]+1, dp[i, j-1]+1, dp[i-1, j-1]+cost)
    return dp[-1, -1]

print("Edit Distance between 'write' and 'right':", edit_distance("write", "right"))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

docs =["Data science is fun", "Machine learning is great", "Apples are tasty", "Bananas are yellow"]
X = TfidfVectorizer().fit_transform(docs)

# Apply K-Means
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10).fit(X)
print("Document Cluster Labels:", kmeans.labels_)
```

B.Given a set of predicted binary labels (y_pred) and their corresponding

```python
import numpy as np

def edit_dist(s1, s2):
    dp = np.zeros((len(s1)+1, len(s2)+1))
    for i in range(len(s1)+1): dp[i,0] = i
    for j in range(len(s2)+1): dp[0,j] = j
    for i in range(1, len(s1)+1):
        for j in range(1, len(s2)+1):
            cost = 0 if s1[i-1] == s2[j-1] else 1
            dp[i,j] = min(dp[i-1,j]+1, dp[i,j-1]+1, dp[i-1,j-1]+cost)
    return dp[-1,-1]

dictionary = ["spelling", "correction", "algorithm", "information", "retrieval", "system"]
input_word = "infmoration"

best_match = min(dictionary, key=lambda w: edit_dist(input_word, w))
print(f"Suggested correction for '{input_word}': {best_match}")
```

A. Implement the Boolean retrieval model for the following corpus. [20]


Document 1: 'this is the first document.'


Document 2: 'this document is the second document.'


Document 3: 'And this is the third one."


Document 4: 'Is this the first document?"


and process the query "first and third".

B. Implement a text summarization algorithm. [20]

```python
# A. Boolean Retrieval Model
docs = {
    1: 'this is the first document.',
    2: 'this document is the second document.',
    3: 'And this is the third one.',
    4: 'Is this the first document?'
}

inverted_index = {}
for doc_id, text in docs.items():
    words = text.lower().replace('.', '').replace('?', '').split()
    for word in words:
        inverted_index.setdefault(word, set()).add(doc_id)

set_first = inverted_index.get('first', set())
set_third = inverted_index.get('third', set())
result = set_first & set_third

print("Inverted Index for 'first':", set_first)
print("Inverted Index for 'third':", set_third)
print("Result for query 'first AND third':", list(result))
```

```python
import networkx as nx
from sklearn.feature_extraction.text import TfidfVectorizer

text = "Natural language processing is a field of computer science. It is related to human–computer interaction. Text summarization is the process of distilling the most important information. Automatic text summarization methods are greatly needed."
sentences = text.split(". ")

matrix = TfidfVectorizer().fit_transform(sentences)

similarity_graph = matrix * matrix.T
scores = nx.pagerank(nx.from_scipy_sparse_array(similarity_graph))

best_sentence_idx = max(scores, key=scores.get)
print("\nOriginal text length:", len(sentences), "sentences")
print("Extracted Summary:", sentences[best_sentence_idx])
```
