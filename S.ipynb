```python
input.jsp code of factorial

@WebMethod(operationName = "checkPrime")
public String checkPrime(@WebParam(name = "n") int n) {
    if (n < 2) return "Not Prime";
    for (int i = 2; i <= n / 2; i++) {
        if (n % i == 0) return "Not Prime";
    }
    return "It is Prime";
}
```
