**Aim** :- ANOVA (Analysis of Variance)
*  Perform one-way ANOVA to compare means across multiple groups.
*  Conduct post-hoc tests to identify significant differences between group means.


## 8. One-Way ANOVA (F-Test)

**Hypothesis:** Is there a significant difference in the 'green_space_ratio' among the different countries?

```python
if not df.empty and df['country'].nunique() >= 2:
    # Prepare the data for ANOVA
    df_anova = df[['green_space_ratio', 'country']]
    groups = pd.unique(df_anova.country.values)
    d_data = {group: df_anova['green_space_ratio'][df_anova.country == group] for group in groups}

    # Perform one-way ANOVA
    # The * operator unpacks the dictionary values into separate arguments
    F, p = stats.f_oneway(*d_data.values())

    print("Hypothesis: Is green space ratio different across countries?")
    print("p-value for significance is: ", p)

    if p < 0.05:
        print("Conclusion: Reject the null hypothesis; there is a significant difference.")
    else:
        print("Conclusion: Accept the null hypothesis.")
```

## 9. Two-Way ANOVA (F-Test)

**Hypothesis:** Does 'avg_income' vary based on an interaction between 'country' and 'happiness_category'?

```python
if not df.empty and df['country'].nunique() >= 2 and df['happiness_category'].nunique() >= 2:
    # Fit the ANOVA model
    # C() indicates that the variable is categorical
    model = ols('avg_income ~ C(country) * C(happiness_category)', data=df).fit()
    
    print(f"Overall model F({model.df_model:.0f}, {model.df_resid:.0f}) = {model.fvalue:.3f}, p = {model.f_pvalue:.4f}")
    
    # Perform two-way ANOVA
    res = sm.stats.anova_lm(model, typ=2)
    print("\nTwo-Way ANOVA Results:")
    print(res)
```
